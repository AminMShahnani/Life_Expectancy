```python
import pandas as pd
df=pd.read_csv('WPP2022.csv')
df.head(15)
```

```python
df=df.drop(df[0:15].index)
df
```

```python
cols=df.iloc[0].to_list()
df.columns=cols
df
```

```python
df=df.set_index('Index')
```

```python
df.columns.to_list()
```

```python
dff=df[['Region, subregion, country or area *','Type','Location code','Parent code','Year','Total Population, as of 1 July (thousands)',
 'Male Population, as of 1 July (thousands)',
 'Female Population, as of 1 July (thousands)','Births (thousands)',
 'Births by women aged 15 to 19 (thousands)','Total Deaths (thousands)','Life Expectancy at Birth, both sexes (years)',
 'Male Life Expectancy at Birth (years)',
 'Female Life Expectancy at Birth (years)']]
```

```python
dff.drop('Index', inplace=True)
type=dff['Type']
```

```python
type
```

```python
dff['Total Population, as of 1 July (thousands)']=dff['Total Population, as of 1 July (thousands)'].replace({' ':''}, regex=True)
```

```python
for col in dff.iloc[:,1:]:
    print(col)
```

```python
for col in dff.iloc[:,1:]:
    #dff[col]=dff[col].replace({' ':''}, regex=True)
    dff[col]= pd.to_numeric(dff[col].replace({' ':''}, regex=True), errors='coerce')
```

```python
dff['Type']=type
dff
```

```python
cond2000=(dff['Year']==2000)
cond2021=(dff['Year']==2021)
new=dff[(cond2000) | (cond2021)]
new
```

```python
dff['prevdif'] = dff.groupby(['Region, subregion, country or area *'])['Life Expectancy at Birth, both sexes (years)'].diff()
dff
```

```python
new['prevdif'] = new.groupby(['Region, subregion, country or area *'])['Life Expectancy at Birth, both sexes (years)'].diff()
new
```
